```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/AKASH HN/Downloads/Dataset/Position_Salaries.csv")
data
```

```python
X = data.iloc[:, 1:2].values
y = data.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
model_lin = LinearRegression()
model_lin.fit(X,y)
```

```python
plt.scatter(X,y, label = 'actual value')
plt.plot(X,model_lin.predict(X),color = 'red',marker = 'o', label = 'Pridecte value')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 4)
X_new = poly.fit_transform(X)
model_ply = LinearRegression()
model_ply.fit(X_new,y)
```

```python
plt.scatter(X, y, label = 'actual value')
plt.plot(X,model_ply.predict(X_new),color = 'red', marker = 'o', label = 'Pridecte value')
plt.legend()
plt.show()
```

#### Predicting new value

```python
print('Prediction of linear model')
model_lin.predict([[5.5]])[0]
```

```python
print('Prediction of Polynomial model')
print(f'{model_ply.predict(poly.fit_transform([[5.5]]))[0]:.2f}')
```
